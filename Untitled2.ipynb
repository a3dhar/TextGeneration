```python
import json, numpy as np, tensorflow as tf, model, sample, encoder
```

```python
def text_model(name, seed, num_samples, batch_size, length, temperature, top_k = 40, model_dir = '/content/gpt-2/models'):
  #assign directory
  model_dir = os.path.expanduser(os.path.expandvars(model_dir))
  #assign batch size if no value given
  if batch_size is None:
    batch_size = 1
  #Check that batch size divides number of samples
  assert num_samples % batch_size == 0

  encode = encoder.get_encoder(name, model_dir)
  
  #get default hyperparameters from pretrained model
  hparams = model.default_hparams()

  with open(os.path.join(model_dir,name,'hparams.json')) as f:
    hparams.override_from_dict(json.load(f))
  
  if length is None:
    length = hparams.n_ctx // 2
  elif length > hparams.n_ctx:
    raise ValueError("cant get samples longer than window dumbass %s" % hparams.n_ctx)
  
  with tf.Session(graph=tf.Graph()) as cur_sesh:
    context = tf.placeholder(tf.int32, [batch_size, None])
    #use seed for repeatability 
    np.random.seed(seed)
    tf.set_random_seed(seed)

    output = sample.sample_sequence(hparams = hparams, length = length, context = context, batch_size = batch_size, temperature = temperature, top_k = top_k)


    saver = tf.train.Saver()
    #save checkpoint
    ckpt = tf.train.latest_checkpoint(os.path.join(model_dir,name))

    saver.restore(cur_sesh, ckpt)

    while True:
      raw = input('Speak to me child ')

      while not raw:
        print('COOOK')
        raw = input('Speak to me child ')
      
      context_token = encode.encode(raw)

      generated = 0

      for _ in range(num_samples // batch_size):
        out = cur_sesh.run(output, feed_dict = {context : [context_token for _ in range(batch_size)]})[:,len(context_token):]
        for i in range(batch_size):
          generated+=1
          text = encode.decode(out[i])
          print("=" * 40 + " SAMPLE " + str(generated)+" "+ "="*40)
          print(text)
      print("=" * 80)

```

```python
text_model('345M',None,1,1,300,1,0,'/content/gpt-2/models')
```
